```python
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv('C:/Users/Carlos/Downloads/Stores.csv')
df
```

```python
df.describe()
```

### Relación del area con los productos disponibles

```python
sns.jointplot(x= 'Store_Area', y= 'Items_Available', data= df, kind= 'reg', color='g')
```

Existe una relación lineal, a mayor area mas cantidad de productos.

### Relación del area con el número de clientes.

```python
sns.jointplot(x= 'Store_Area', y= 'Daily_Customer_Count', data= df, kind='reg', color='g')
```

El area no tiene relación con el número de clientes.

### Relación del area con las ventas

```python
sns.jointplot(x= 'Store_Area', y= 'Store_Sales', data= df, kind='reg', color='g')
```

No existe relación entre el area y las ventas

### Relación del número de clientes con los productos disponibles

```python
sns.jointplot(x= 'Items_Available', y= 'Daily_Customer_Count', data= df, kind='reg', color= 'g')
```

No hay relación entre estas dos variables

### Relación del número de productos y las ventas

```python
sns.jointplot(x= 'Items_Available', y= 'Store_Sales', data= df, kind= 'reg', color= 'g')
```

Mayor numero de productos disponibles no quiere decir mayores ventas

### Relación del número de clientes con las ventas

```python
sns.jointplot(x= 'Daily_Customer_Count', y= 'Store_Sales', data= df, kind='reg', color='g')
```

Mayor numero de clientes no se ve reflejado en las ventas
